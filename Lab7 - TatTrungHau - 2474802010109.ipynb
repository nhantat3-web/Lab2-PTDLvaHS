```
# =========================
# BÀI 3: PHÂN TÍCH TẦN SỐ CỦA CÁC TỪ
# =========================

from nltk.corpus import gutenberg, stopwords
import nltk
import string

macbeth = gutenberg.words('shakespeare-macbeth.txt')

# Tần số từ ban đầu
fd = nltk.FreqDist(macbeth)

print("===== 10 TỪ XUẤT HIỆN NHIỀU NHẤT =====")
print(fd.most_common(10))

# Stopwords
sw = set(stopwords.words('english'))

print("\n===== SỐ LƯỢNG STOPWORDS TIẾNG ANH =====")
print(len(sw))

print("\n===== 10 STOPWORDS ĐẦU TIÊN =====")
print(list(sw)[:10])

# Bỏ stopwords
macbeth_filtered = [
    w for w in macbeth
    if w.lower() not in sw
]

print("\n===== ĐỘ DÀI SAU KHI BỎ STOPWORDS =====")
print(len(macbeth_filtered))

fd = nltk.FreqDist(macbeth_filtered)

print("\n===== 10 TỪ PHỔ BIẾN SAU KHI BỎ STOPWORDS =====")
print(fd.most_common(10))

# Bỏ stopwords và dấu câu
punctuation = set(string.punctuation)

macbeth_filtered2 = [
    w.lower() for w in macbeth
    if w.lower() not in sw and w.lower() not in punctuation
]

fd = nltk.FreqDist(macbeth_filtered2)

print("\n===== 10 TỪ PHỔ BIẾN SAU KHI BỎ STOPWORDS VÀ DẤU CÂU =====")
print(fd.most_common(10))
```

```
# =========================
# BÀI 4: LỰA CHỌN CÁC TỪ TRONG VĂN BẢN
# =========================

from nltk.corpus import gutenberg

macbeth = gutenberg.words('shakespeare-macbeth.txt')

# Rút trích các từ có độ dài lớn hơn 12 ký tự
long_words = [
    w for w in macbeth
    if len(w) > 12
]

print("===== CÁC TỪ DÀI HƠN 12 KÝ TỰ =====")
print(sorted(set(long_words)))

# Rút trích các từ có chứa chuỗi 'ious'
ious_words = [
    w for w in macbeth
    if 'ious' in w
]

ious_words = set(ious_words)

print("\n===== CÁC TỪ CÓ CHỨA CHUỖI 'ious' =====")
print(sorted(ious_words))
```

```
# =========================
# BÀI 5: BIGRAMS VÀ COLLOCATIONS
# =========================

from nltk.corpus import gutenberg, stopwords
import nltk
import string

macbeth = gutenberg.words('shakespeare-macbeth.txt')

sw = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Lọc stopwords và dấu câu
macbeth_filtered2 = [
    w.lower() for w in macbeth
    if w.lower() not in sw and w.lower() not in punctuation
]

# Bigram
bgrms = nltk.FreqDist(
    nltk.bigrams(macbeth_filtered2)
)

print("===== 15 BIGRAM PHỔ BIẾN NHẤT =====")
print(bgrms.most_common(15))

# Trigram
tgrms = nltk.FreqDist(
    nltk.trigrams(macbeth_filtered2)
)

print("\n===== 10 TRIGRAM PHỔ BIẾN NHẤT =====")
print(tgrms.most_common(10))
```

```
# =========================
# BÀI 8: PHÂN TÍCH CẢM XÚC NGƯỜI DÙNG
# =========================

import nltk
import random
from nltk.corpus import movie_reviews

print("===== PHÂN TÍCH CẢM XÚC MOVIE REVIEWS =====")

documents = [
    (list(movie_reviews.words(fileid)), category)
    for category in movie_reviews.categories()
    for fileid in movie_reviews.fileids(category)
]

random.shuffle(documents)

print("\n===== REVIEW ĐẦU TIÊN =====")
first_review = ' '.join(documents[0][0])
print(first_review[:1000])

print("\n===== NHÃN CẢM XÚC CỦA REVIEW ĐẦU TIÊN =====")
print(documents[0][1])

all_words = nltk.FreqDist(
    w.lower() for w in movie_reviews.words()
)

# Lấy 2000 từ phổ biến nhất để chạy nhanh hơn
word_features = list(all_words)[:2000]

print("\n===== SỐ LƯỢNG TỪ ĐẶC TRƯNG =====")
print(len(word_features))


def document_features(document, word_features):
    document_words = set(document)
    features = {}

    for word in word_features:
        features[word] = word in document_words

    return features


featuresets = [
    (document_features(d, word_features), c)
    for (d, c) in documents
]

print("\n===== SỐ LƯỢNG FEATURESETS =====")
print(len(featuresets))

# Chia tập train và test
train_set = featuresets[1500:]
test_set = featuresets[:500]

classifier = nltk.NaiveBayesClassifier.train(train_set)

accuracy = nltk.classify.accuracy(classifier, test_set)

print("\n===== ĐỘ CHÍNH XÁC CỦA MÔ HÌNH =====")
print(accuracy)

print("\n===== 10 ĐẶC TRƯNG QUAN TRỌNG NHẤT =====")
classifier.show_most_informative_features(10)
```

```
# =========================
# BÀI 9.4: LOẠI BỎ STOPWORDS
# =========================

import nltk
from nltk.corpus import stopwords

sample_text = "This is a simple example showing how to remove stopwords from a text."

words = nltk.word_tokenize(sample_text)

sw = set(stopwords.words('english'))

filtered_words = [
    word for word in words
    if word.lower() not in sw
]

print("Văn bản ban đầu:")
print(words)

print("\nVăn bản sau khi bỏ stopwords:")
print(filtered_words)
```

```
# =========================
# BÀI 9.5: BỎ STOPWORDS VÀ DẤU CÂU
# =========================

import nltk
import string
from nltk.corpus import stopwords

sample_text = "This is a simple example, showing how to remove stopwords from a text."

words = nltk.word_tokenize(sample_text)

sw = set(stopwords.words('english'))
punctuation = set(string.punctuation)

filtered_words = [
    word for word in words
    if word.lower() not in sw and word not in punctuation
]

print("Văn bản sau khi bỏ stopwords và dấu câu:")
print(filtered_words)
```

```
# =========================
# BÀI 9.6: TÌM ĐỊNH NGHĨA VÀ VÍ DỤ BẰNG WORDNET
# =========================

from nltk.corpus import wordnet as wn

word = "computer"

synsets = wn.synsets(word)

print("Định nghĩa và ví dụ của từ:", word)

for syn in synsets[:3]:
    print("\nTên synset:", syn.name())
    print("Định nghĩa:", syn.definition())
    print("Ví dụ:", syn.examples())
```

```
# =========================
# BÀI 9.7: TỪ ĐỒNG NGHĨA VÀ TRÁI NGHĨA
# =========================

from nltk.corpus import wordnet as wn

synonyms = []
antonyms = []

for syn in wn.synsets("good"):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())

        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print("Từ đồng nghĩa của good:")
print(set(synonyms))

print("\nTừ trái nghĩa của good:")
print(set(antonyms))
```
