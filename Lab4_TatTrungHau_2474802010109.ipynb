# AI

Vấn đề 1:

```
import pandas as pd

column_names = [
    'Id',
    'Name',
    'Age',
    'Weight',
    'm0006',
    'm0612',
    'm1218',
    'f0006',
    'f0612',
    'f1218'
]

df = pd.read_excel(
    "patient_heart_rate.xlsx",
    header=None,
    names=column_names
)

print(df.head())
print(df.columns)
```

Vấn đề 2:

```
df[['Firstname','Lastname']] = (
    df['Name']
    .astype(str)
    .str.split(' ', n=1, expand=True)
)

df = df.drop('Name', axis=1)

print(df.head())
```

Vấn đề 3:


```
def convert_weight(x):

    if pd.isna(x):
        return None

    x = str(x)

    if "lbs" in x:
        value = float(x.replace("lbs", ""))
        return round(value / 2.2, 2)

    if "kgs" in x:
        return float(x.replace("kgs", ""))

    return x

df["Weight"] = df["Weight"].apply(convert_weight)

print(df)
print(df.head())
```

Vấn đề 4

```
df = df.dropna(how="all")

print(df)
```

Vấn đề 5:


```
df = df.drop_duplicates(
    subset=['Firstname','Lastname','Age','Weight']
)

print(df)
```

Vấn đề 6

```
df["Firstname"] = (
    df["Firstname"]
    .astype(str)
    .str.encode("ascii","ignore")
    .str.decode("ascii")
)

df["Lastname"] = (
    df["Lastname"]
    .astype(str)
    .str.encode("ascii","ignore")
    .str.decode("ascii")
)

print(df)
```

Vấn đề 7:


```
print(df[['Age','Weight']].isnull().sum())

df = df.dropna(
    subset=['Age','Weight'],
    how='all'
)

mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

mean_weight = df['Weight'].mean()
df['Weight'] = df['Weight'].fillna(mean_weight)

print(df[['Age','Weight']].isnull().sum())
print(df)
```

Vấn đề 8:

```
df = pd.melt(
    df,
    id_vars=[
        'Id',
        'Age',
        'Weight',
        'Firstname',
        'Lastname'
    ],
    value_name='PulseRate',
    var_name='sex_and_time'
)

tmp_df = df['sex_and_time'].str.extract(
    r'(\D)(\d{2})(\d{2})',
    expand=True
)

tmp_df.columns = [
    'Sex',
    'hours_lower',
    'hours_upper'
]

tmp_df['Time'] = (
    tmp_df['hours_lower']
    + '-'
    + tmp_df['hours_upper']
)

df = pd.concat(
    [df,tmp_df],
    axis=1
)

df = df.drop(
    [
        'sex_and_time',
        'hours_lower',
        'hours_upper'
    ],
    axis=1
)

print(df)
```

Vấn đề 9:


```
df['PulseRate'] = pd.to_numeric(
    df['PulseRate'],
    errors='coerce'
)

df['PulseRate'] = (
    df.groupby('Id')['PulseRate']
    .transform(
        lambda x: x.fillna(x.mean())
    )
)

df['PulseRate'] = (
    df.groupby('Sex')['PulseRate']
    .transform(
        lambda x: x.fillna(x.mean())
    )
)

global_mean = df['PulseRate'].mean()

df['PulseRate'] = (
    df['PulseRate']
    .fillna(global_mean)
)

print(df)
```

vấn đề 10:


```
df = df.reset_index(drop=True)

df.to_csv(
    "patient_heart_rate_clean.csv",
    index=False
)

print(df.head())

print("Đã lưu file patient_heart_rate_clean.csv")
print(df.info())
print(df.isnull().sum())
```
