# LAB 2 - Bài thực hành trình bày dữ liệu

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from pandas.plotting import scatter_matrix

FILE_PATH = "processed_dulieuxettuyendaihoc.csv.xlsx"

# Đọc file Excel
df = pd.read_excel(FILE_PATH, sheet_name=0, header=1)

# Chuẩn hóa kiểu dữ liệu
for col in df.columns:
    if col not in ["GT", "KV", "KT", "XL1", "XL2", "XL3"]:
        df[col] = pd.to_numeric(df[col], errors="ignore")

print("Kích thước dữ liệu:", df.shape)
display(df.head())
display(df.info())
```

# PHẦN 1: THỐNG KÊ DỮ LIỆU

# Câu 1. Sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần

```
cau1 = df.sort_values(by="DH1", ascending=True)
display(cau1)
```

# Câu 2. Sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính

```
cau2 = df.sort_values(by=["GT", "DH2"], ascending=[True, True])
display(cau2[["STT", "GT", "DH1", "DH2", "DH3", "KT", "KV", "DT"]])
```

```
# Hàm thống kê dùng cho pivot-table
def thong_ke_nhom(data, group_cols, value_col="DH1"):
    kq = data.groupby(group_cols)[value_col].agg(
        count="count",
        sum="sum",
        mean="mean",
        median="median",
        min="min",
        max="max",
        std="std",
        Q1=lambda x: x.quantile(0.25),
        Q2=lambda x: x.quantile(0.50),
        Q3=lambda x: x.quantile(0.75)
    )
    return kq.round(3)

# Câu 3. Pivot-table thống kê DH1 theo KT
pivot_cau3 = thong_ke_nhom(df, ["KT"], "DH1")
display(pivot_cau3)
```

# Câu 4. Pivot-table thống kê DH1 theo KT và KV

```
pivot_cau4 = thong_ke_nhom(df, ["KT", "KV"], "DH1")
display(pivot_cau4)
```

# Câu 5. Pivot-table thống kê DH1 theo KT, KV và DT

```
pivot_cau5 = thong_ke_nhom(df, ["KT", "KV", "DT"], "DH1")
display(pivot_cau5)
```

# Câu 1. Trình bày biến GT: bảng tần số, tần suất, biểu đồ cột, biểu đồ tròn

```
freq_GT = df["GT"].value_counts().rename("Tần số")
rate_GT = (df["GT"].value_counts(normalize=True) * 100).round(2).rename("Tần suất (%)")
bang_GT = pd.concat([freq_GT, rate_GT], axis=1)
display(bang_GT)

plt.figure(figsize=(6,4))
freq_GT.plot(kind="bar")
plt.title("Biểu đồ tần số biến GT")
plt.xlabel("Giới tính")
plt.ylabel("Tần số")
plt.xticks(rotation=0)
plt.show()

plt.figure(figsize=(5,5))
freq_GT.plot(kind="pie", autopct="%1.1f%%")
plt.title("Biểu đồ tần suất biến GT")
plt.ylabel("")
plt.show()
```

```
# thỏa điều kiện DH1 >= 5.0, DH2 >= 4.0, DH3 >= 4.0
# Trong file này mã DT xuất hiện nhiều nhất được xem là nhóm Kinh.
KINH_CODE = df["DT"].mode()[0]

loc_cau4 = df[
    (df["GT"] == "M") &
    (df["DT"] == KINH_CODE) &
    (df["DH1"] >= 5.0) &
    (df["DH2"] >= 4.0) &
    (df["DH3"] >= 4.0)
]

bang_KV = loc_cau4["KV"].value_counts().sort_index()

display(loc_cau4[["STT","GT","DT","KV","DH1","DH2","DH3"]])

display(pd.DataFrame({
    "Tần số": bang_KV,
    "Tần suất (%)": (bang_KV / bang_KV.sum() * 100).round(2)
}))
```

```
bins = [0, 5, 7, 8, 10]
labels = ["Kem", "Trungbinh", "Kha", "Gioi"]

df["phanlopt1"] = pd.cut(
    df["T1"],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)

# Gộp trường hợp T1 = 10 vào nhóm Giỏi
df.loc[df["T1"] == 10, "phanlopt1"] = "Gioi"

display(df[["T1", "phanlopt1"]].head(20))
display(df["phanlopt1"].value_counts())
```

# PHẦN 5: MÔ TẢ DỮ LIỆU VÀ KHẢO SÁT DẠNG PHÂN PHỐI


# Câu 1. Mô tả và khảo sát phân phối cho biến T1

```
x = df["T1"].dropna()

mo_ta_T1 = pd.Series({
    "count": x.count(),
    "mean": x.mean(),
    "std": x.std(),
    "min": x.min(),
    "Q1": x.quantile(0.25),
    "median/Q2": x.median(),
    "Q3": x.quantile(0.75),
    "max": x.max(),
    "skewness": x.skew(),
    "kurtosis": x.kurtosis()
}).round(3)

display(mo_ta_T1)

Q1 = x.quantile(0.25)
Q2 = x.quantile(0.50)
Q3 = x.quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
outliers = x[(x < lower_fence) | (x > upper_fence)]

box_10 = pd.Series({
    "Min": x.min(),
    "Q1": Q1,
    "Median/Q2": Q2,
    "Q3": Q3,
    "Max": x.max(),
    "IQR": IQR,
    "Lower fence": lower_fence,
    "Upper fence": upper_fence,
    "Mean": x.mean(),
    "Số ngoại lệ": outliers.count()
}).round(3)

display(box_10)

plt.figure(figsize=(6,5))
plt.boxplot(x, vert=True)
plt.title("Box-plot của T1")
plt.ylabel("T1")
plt.show()

plt.figure(figsize=(7,5))
plt.hist(x, bins=10, edgecolor="black")
plt.title("Histogram của T1")
plt.xlabel("T1")
plt.ylabel("Tần số")
plt.show()

plt.figure(figsize=(6,5))
stats.probplot(x, dist="norm", plot=plt)
plt.title("QQ-plot kiểm chứng phân phối chuẩn của T1")
plt.show()

print("Nhận xét nhanh:")
print("- Skewness > 0: lệch phải, Skewness < 0: lệch trái, gần 0: gần đối xứng.")
print("- Kurtosis > 0: nhọn hơn chuẩn, Kurtosis < 0: bẹt hơn chuẩn.")
print("- Nếu điểm QQ-plot nằm gần đường thẳng thì dữ liệu gần phân phối chuẩn.")
```

# Câu 2. Mô tả và khảo sát phân phối T1 trên từng nhóm phanlopt1

```
display(df.groupby("phanlopt1")["T1"].describe().round(3))

groups = [df.loc[df["phanlopt1"] == label, "T1"].dropna() for label in labels]

plt.figure(figsize=(8,5))
plt.boxplot(groups, labels=labels)
plt.title("Box-plot T1 theo phanlopt1")
plt.xlabel("phanlopt1")
plt.ylabel("T1")
plt.show()

for label in labels:
    data = df.loc[df["phanlopt1"] == label, "T1"].dropna()
    plt.figure(figsize=(6,4))
    plt.hist(data, bins=6, edgecolor="black")
    plt.title(f"Histogram T1 - nhóm {label}")
    plt.xlabel("T1")
    plt.ylabel("Tần số")
    plt.show()

for label in labels:
    data = df.loc[df["phanlopt1"] == label, "T1"].dropna()
    if len(data) >= 3:
        plt.figure(figsize=(6,5))
        stats.probplot(data, dist="norm", plot=plt)
        plt.title(f"QQ-plot T1 - nhóm {label}")
        plt.show()
    else:
        print(f"Nhóm {label} có quá ít dữ liệu để vẽ QQ-plot.")
```

# Câu 3. Khảo sát tương quan giữa DH1 và T1

```
cov_DH1_T1 = df[["DH1", "T1"]].cov()
corr_DH1_T1 = df[["DH1", "T1"]].corr()

print("Ma trận hiệp phương sai:")
display(cov_DH1_T1.round(3))

print("Ma trận tương quan:")
display(corr_DH1_T1.round(3))

plt.figure(figsize=(6,5))
plt.scatter(df["T1"], df["DH1"])
plt.title("Scatter: DH1 theo T1")
plt.xlabel("T1")
plt.ylabel("DH1")
plt.grid(True, alpha=0.3)
plt.show()
```

# Câu 4. Khảo sát tương quan giữa DH1 và T1 trên từng nhóm khu vực

```
for kv in sorted(df["KV"].dropna().unique()):
    sub = df[df["KV"] == kv]
    print(f"KV = {kv}")
    display(sub[["DH1", "T1"]].corr().round(3))

    plt.figure(figsize=(6,5))
    plt.scatter(sub["T1"], sub["DH1"])
    plt.title(f"Scatter DH1 theo T1 - KV {kv}")
    plt.xlabel("T1")
    plt.ylabel("DH1")
    plt.grid(True, alpha=0.3)
    plt.show()
```
